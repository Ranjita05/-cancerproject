##splitting data

```python
#splitting the data
X= c1.iloc[:,2:31].values
Y= c1.iloc[:,1].values
```

##training model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.25, random_state= 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test= sc.fit_transform(X_test)
```

```python
def models(X_train, Y_train):
    
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train, Y_train)
    
    #decision tree
    from sklearn.tree import DecisionTreeClassifier
    tree= DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(X_train, Y_train)
    
    #random forest classifier
    from sklearn.ensemble import RandomForestClassifier
    forest= RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state=0)
    forest.fit(X_train, Y_train)
    
    print('[0]Logistic Regression Training Accuracy:', log.score(X_train, Y_train),
          '[1]Decision Tree Classifier Training Accuracy:', log.score(X_train, Y_train),
          '[2]Random Forest Classifier Training Accuracy:', log.score(X_train, Y_train)
         )
    return log, tree, forest
```

```python
model=models(X_train, Y_train)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix

for i in range( len(model) ):
    print('Model', i)
    cn= confusion_matrix(Y_test, model[i].predict(X_test))

    TP= cn[0][0]
    TN= cn[1][1]
    FN= cn[1][0]
    FP= cn[0][1]

    print(cn)
    print('Testing Accuracy=', (TP + TN)/ (TP + TN + FN + FP))
    print()
```

```python
##another method
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range( len(model) ):
    print('Model',i)
    print(classification_report(Y_test, model[i].predict(X_test)))
    print(accuracy_score(Y_test,model[0].predict(X_test)))
    print()
```
